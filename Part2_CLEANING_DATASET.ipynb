# Part 2: Data Cleaning

```python
import os
import matplotlib.pyplot as plt
from PIL import Image
path = r"data/CORROSION"
imgs = os.listdir(path)
c1=0
print(len(imgs))
for img in imgs:
    img_path = path +'/'+ img
    try:
        plt.imshow(Image.open(img_path))
    except:
        c1 +=1
        os.remove(img_path)
print(c1)
imgs = os.listdir(path)
print(len(imgs))
```
